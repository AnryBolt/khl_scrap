```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import pandas as pd
```

```python
url_home = "https://www.khl.ru/stat/teams/1217/home"
url_road = "https://www.khl.ru/stat/teams/1217/road"
```

```python
def get_html_with_selenium(url, name):
    options = webdriver.FirefoxOptions()
    options.set_preference("general.useragent.override", "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36")

    try:
        driver = webdriver.Firefox(options=options)
        driver.get(url=url)
        time.sleep(5)
    
        with open(name, "w", encoding='UTF-8') as file:
            file.write(driver.page_source)
    
    except Exception as ex:
        print(ex)
    finally:
        driver.close()
        driver.quit()
```

```python
def get_table_khl(url, name):

    get_html_with_selenium(url, name)
    html_name = name + ".html"
    with open(html_name, encoding='UTF-8') as file:
        src = file.read()

    soup = BeautifulSoup(src, "lxml")
    sp = soup.find("table").\
        text.strip().\
            split("\n")

    sp = list(filter(lambda x: x.strip() != "", sp))

    n = 0
    sp2 = []
    for i in range(int(len(sp)/19)):
        sp2.append(sp[n :n + 19])
        n = n + 19
    df = pd.DataFrame(sp2[1:], columns=sp2[0])
    return df
```

```python
get_table_khl(url_home, "home")
```
