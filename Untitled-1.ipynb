```python
def get_html_with_selenium(url, name): # функция получения html страницы с имененм name и по адресу url
    options = webdriver.FirefoxOptions()
    options.set_preference("general.useragent.override", "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36")

    try:
        driver = webdriver.Firefox(options=options)
        driver.get(url=url)
        time.sleep(5)
    
        with open(name, "w", encoding='UTF-8') as file:
            file.write(driver.page_source)
    
    except Exception as ex:
        print(ex)
    finally:
        driver.close()
        driver.quit()
```

```python
def get_table_khl(url, name): # Функция получения таблицы khl  с нужной страницы и пгенерация html страницы с этой таблицы именем "name.html"
    html_name = name + ".html"
    get_html_with_selenium(url, html_name)
    with open(html_name, encoding='UTF-8') as file:
        src = file.read()

    soup = BeautifulSoup(src, "lxml")
    sp = soup.find("table").\
        text.strip().\
            split("\n")

    sp = list(filter(lambda x: x.strip() != "", sp))

    n = 0
    sp2 = []
    for i in range(int(len(sp)/19)):
        sp2.append(sp[n :n + 19])
        n = n + 19
    df = pd.DataFrame(sp2[1:], columns=sp2[0])
    df.index = df['Клуб']
    del df['Клуб']
    df = df.astype(int)
    df = df.reset_index()
    return df
```

```python
def get_calendar_khl(name): # Функция получения календаря khl  с нужной страницы и генерация html страницы с этой таблицы именем "name.html"
    html_name = name + ".html"

    get_html_with_selenium("https://www.khl.ru/calendar/", name)
    with open(html_name, encoding='UTF-8') as file:
        src = file.read()

    soup = BeautifulSoup(src, "lxml")

    com = soup.findAll("p",class_="card-game__club-name roboto roboto-bold roboto-lg color-dark link")
    com_text = []
    for i in com:
        com_text.append(i.text)
    com_home = com_text[::2][:186] #1 домашние команды
    com_road = com_text[1::2][:186] #2 гостевые команды

    number = soup.findAll("p",class_="card-game__center-number roboto-condensed roboto-normal roboto-lg color-semiSecondary")
    number_text = [] #3 Номер матча
    for i in number:
        number_text.append(int(i.text.replace("№ ","")))
    number_text = number_text[:186]

    time = soup.findAll("p",class_="card-game__center-time roboto-condensed roboto-bold title-md color-dark")
    time_text = [] #4 Время
    for i in time:
        time_text.append(i.text.replace("мск\n", ""))

    dates = soup.findAll("div",class_="calendary-body__wrap-header roboto roboto-bold roboto-xxl color-black")
    dates_text = [] # Даты
    for i in dates:
        dates_text.append(i.text.strip().replace("\n\n\n",", "))

    tables = soup.findAll("div",class_="card-games card-games_list card-games-style-type fade")
    n_in_tables = [] # кол-во игр в таблице
    for i in tables:
        n_in_tables.append(len(i.findAll("div",class_="card-game card-game--calendar")))

    date_text = [] #5 Дата
    for i, z in enumerate(dates_text):
        date_text.extend([z] * n_in_tables[i])

    df = pd.DataFrame([com_home, com_road, number_text, time_text, date_text]).transpose()
    df.columns=['home','road', '№_games', 'time', 'date']
    df['Game'] = df['home'] + ' - ' + df['road']
    del df['home']
    del df['road']
    return df
```

```python
sl = {# словарь определенных страниц khl
    "home" : "https://www.khl.ru/stat/teams/1217/home",\
      "road" : "https://www.khl.ru/stat/teams/1217/road"\
        }
```

Получение словаря таблиц khl

```python
sl_tables = {}
for i in sl:
    sl_tables[i] = get_table_khl(sl[i], i)
```

```python
sl_tables.keys()
```

```python
home = sl_tables['home'].sort_values(['Ибз', 'ЗШ'], ascending = [True, False])
home_best = home[home['Ибз'] <= 1][['Клуб', 'Ибз', 'ЗШ']].reset_index()
del home_best['index']
home_best
```

```python
road = sl_tables['road'].sort_values(['Ибп', 'ПШ'], ascending = [True, False])
road_bad = road[road['Ибп'] <= 1][['Клуб', 'Ибп', 'ПШ']].reset_index()
del road_bad['index']
road_bad
```

```python
cross = home_best.merge(road_bad, how="cross")
cross['Игра'] = cross['Клуб_x'] + ' - ' + cross['Клуб_y']
```

Получим календарь

```python
calendar = get_calendar_khl('calendar_khl')
```
